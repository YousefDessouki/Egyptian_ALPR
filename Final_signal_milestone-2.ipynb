```python
# Load a model
model = YOLO('yolov8n.pt')  # load a pretrained model (recommended for training)

# Load the model
model = YOLO('/content/runs/detect/train8/weights/best.pt')

# Predict on an image
result = model.predict('/content/Milestone-3-test-images/0210.jpg', save=True, conf=0.1, hide_conf=True,agnostic_nms=True, save_crop=True)
```

```python
import torch
import numpy as np

# Assuming 'result' contains the bounding boxes
boxes = result[0]

# Convert the bounding boxes to a NumPy array
boxes = boxes.detach().cpu().numpy()

# Initialize a list to store the bounding box coordinates
bounding_boxes = []

# Iterate over each bounding box
for i, box in enumerate(boxes):
    # Access the position information for each box
    x_min, y_min, x_max, y_max = box[:4]

    # Store the bounding box coordinates in a list
    bounding_boxes.append([x_min, y_min, x_max, y_max])

    # Print the position information for each box
    print(f"Bounding box {i + 1} coordinates: ({x_min}, {y_min}), ({x_max}, {y_max})")

# Convert the list of bounding box coordinates to a 2D NumPy array
bounding_boxes_array = np.array(bounding_boxes)

# Print the 2D array of bounding box coordinates
print("Bounding boxes array:")
print(bounding_boxes_array)

```

```python
import cv2
import os

def extract_images(image_path, bounding_boxes_array):
    """
    Extract images based on the bounding box coordinates.

    Args:
    image_path (str): Path to the original image.
    bounding_boxes_array (numpy.ndarray): 2D array of bounding box coordinates where each row represents a bounding box in the format [x_min, y_min, x_max, y_max].

    Returns:
    None
    """
    # Load the original image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Image at {image_path} could not be loaded.")
        return

    # Create a directory to save cropped images
    output_dir = os.path.splitext(image_path)[0] + "_cropped"
    os.makedirs(output_dir, exist_ok=True)

    # Iterate over each bounding box
    for i, box in enumerate(bounding_boxes_array):
        # Extract bounding box coordinates
        x_min = int(box[0])
        y_min = int(box[1])
        x_max = int(box[2])
        y_max = int(box[3])

        # Crop the region of interest from the original image
        cropped_image = image[y_min:y_max, x_min:x_max]

        # Save the cropped image
        output_path = os.path.join(output_dir, f"box_{i + 1}.jpg")
        cv2.imwrite(output_path, cropped_image)
        print(f"Box {i + 1} saved to {output_path}")

# Example usage
image_path = '/content/runs/detect/predict8/0210.jpg'  # Replace with your image path
extract_images(image_path, bounding_boxes)
```
